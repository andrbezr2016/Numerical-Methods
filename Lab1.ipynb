```julia
using PyPlot
E = 0.001;
n = 1;
```

```julia
#1 Опеделение an:
function a(n)
    return an = 36 / (11(n^2+5n+4))
end
```

```julia
#1 Вычисление Sn:
function S(n)
    Sn = 0
    for i in 1:n
        Sn += a(i)
    end
    return Sn
end
```

```julia
#2 Вычисление L=Lim(Sn):
while(abs((S(n) - S(n-1))) > E)
    n+=1
end
println("n=", n, "\tL=", a(n))
```

```julia
println("Number: ",n , "\tLim(S$n) = ", S(n), "\ta$n = ",a(n))
```

```julia
println("S$n= ",S(n),"\tS$(n-1)= ",S(n-1),"\nS$n - S$(n-1)= ", S(n)-S(n-1))
```

```julia
#3 график a(n):
x = 1:n;
z = a.(x);
plot(x,z,color="blue")
```

```julia
#3 график S(n):
y = S.(x);
plot(x,y,color="red")
```
